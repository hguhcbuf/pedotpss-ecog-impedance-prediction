```python
import os
import cv2
import matplotlib.pyplot as plt

from pipeline import run_full_pipeline
```

```python
count = 1
image_dir = "sample_images"

for file in os.listdir(image_dir):

    print("sample {} : {}".format(count, file))
    file_path = os.path.join(image_dir, file)

    try:
        run_full_pipeline(
            file_path,
            extra_space=10,
            prominence=(0.7, 50),
            gaussian_smooth=30,
            gaussian_sigma=8
        )

    except Exception as e:
        img = cv2.imread(file_path)

        if img is not None:
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            plt.imshow(img_rgb)
            plt.axis("off")
            plt.title("Invalid image for prediction")
            plt.show()
        else:
            print("Could not read image file.")

        print("-" * 40)

        count += 1
        continue

    print("-" * 40)
    count += 1
```
